### user reviews file (yelp_academic_dataset_review.json)
Since the user reviews file is huge (more than 5 GB of data), we can read the file lines as strings to get row count, provided each line represents a row of data.

```python
reviews = f'{dataset_path}/yelp_academic_dataset_review.json'

# Sum up all lines to get row count
with open(reviews, 'r') as file:
    row_count = sum(1 for line in file)

print(f'{row_count} rows in reviews file')
```

```python
# Load a few rows from user reviews
reviews_iter = pd.read_json(f'{dataset_path}/yelp_academic_dataset_review.json', lines=True, chunksize=1000)
reviews_df = next(reviews_iter)
reviews_df.shape
```

```python
print(f'{reviews_df.shape[1]} columns')
```

```python
reviews_df.head()
```

#### Finally, how many users are in the Yelp dataset

### users file (yelp_academic_dataset_user.json)

```python
users = f'{dataset_path}/yelp_academic_dataset_user.json'

# Sum up all lines to get row count
with open(users, 'r') as file:
    row_count = sum(1 for line in file)

print(f'{row_count} rows in users file')
```

```python
# Load a few rows from users file
users_iter = pd.read_json(f'{dataset_path}/yelp_academic_dataset_user.json', lines=True, chunksize=1000)
users_df = next(users_iter)
users_df.shape
```
